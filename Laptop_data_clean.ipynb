# **Data Assessing**
## Quality Issue

1. **model -** some brands name written differently. `consistency`
2.   **price -** used special character and comma `validity`
3.   **price -** some laptop's price are very cheap.`accuracy`
4.  **processor -** performance code are missing (15,415,450,522) `validity`
5.  **processor -** Gen are missing(22,65,74,77,78,80,92,415,450,455,522) `validity`
6.  **Graphic Card -** wrong value present in graphic card (15,21,279,286,331,338,349,407,421,425,443,448,449,450,454,456,457,458,460,461,490,496,510,528,532,533) `validity`
7.  **core -** some wrong value present in core (15,21) `validity`
8.  **core -** missing core (149,375,495) `validity`
9.  **core -** missing thread (415,424,450,467,522) `validity`
10.  **display -** wrong value (15,21,279,284,286,331,338,349,407,421,425,443,448,449,450,454,456,457,458,460,461,479,490,496,510,528,532,533) `validity`
11.  **ram -** wrong value (21) `validity`
12.  **os -** wrong value (15,21,57,153,219,265,279,286,331,338,349,407,421,425,443,448,449,450,454,456,457,458,460,461,479,490,496,510,528,531,532,533) `validity`
13.  **Internal storage -** wrong value (15,21,479) `validity`


# **Tidiness Issue**


1.   **model -** split into one column (brand name)
2.   **processor -** split into 4 column (processor gen, processor name, processor code, processor name codei5...)
3.  **graphic card -** split into 3 columns (graphic card storage, graphic card name, graphic card code)
4.  **display -** split into 3 columns (display size in inches, display size in pixel, touch screen)
5.  **ram -** split into 2 columns (ram storage, ram type)
6.  **internal storage -** split into 2 columns (storage in GB, type of internal storage)
